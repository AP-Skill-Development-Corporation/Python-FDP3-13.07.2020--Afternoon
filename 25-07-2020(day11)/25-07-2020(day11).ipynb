- **loc and iloc**

- loc get rows or columns with particular labels from the index

- iloc gets rows or columns at particular positions in the index(it takes only integers)

```python
a.loc[1,["name"]] # accessing particular data
```

```python
a.iloc[1,0],a.iloc[3,0]
```

```python
a.iloc[0]
```
