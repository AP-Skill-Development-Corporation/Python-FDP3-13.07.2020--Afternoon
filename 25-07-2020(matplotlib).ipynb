```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,100,100)
y = np.random.randint(1,100,100)
plt.scatter(x,y,color='y')
plt.title('xy graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x = [1,2,3,4]
y = [4,2,5,1]
plt.plot(x,y,'g')
plt.plot(y,x,'b')
plt.legend(['team-b','team-a'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('xy graph')
plt.show()
```

```python
x = ['1996','1997','1998','1999']
y = [33,45,34,56]
plt.bar(x,y)
plt.show()
```

```python
a = ['c','a','d','a','c','a']
plt.hist(a)
plt.show()
```
